## Question 9 (Easy)

**Title:** Count Even and Odd Numbers

### Problem Statement:
Take N as input and count how many even and odd numbers are there between 1 and N.

### Approach:
Use two counters and check with modulo operator.


```python
# write your code here

N=int(input("Enter the number: "))

count_even=0
count_odd=0
for i in range(1,N):
  if i%2==0:
    count_even=count_even+1
  else:
    count_odd=count_odd+1

print(f' Count of even {count_even}')
print(f'Count of odd {count_odd}')

print(f'Total Count: {count_odd+count_even}')

```

## Question 16 (Hard)

**Title:** Check Prime Number

### Problem Statement:
Take a number as input and check whether it is prime or not.

### Approach:
Loop from 2 to N-1 and use `break` if divisible.


```python
# write your code here
# Take input
N = int(input("Enter a number: "))

# Prime check

if N <= 1:
    print(N, "is not a prime number")
else:
    for i in range(2, N):
        if N % i == 0:
            print(N, "is not a prime number")
            break
    else:
        print(N, "is a prime number")
```
