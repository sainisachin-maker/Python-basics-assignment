Q5. Reverse a number using while loop.

Approach:
- Extract last digit
- Build reversed number

```python
#write answer here
num=int(input("Enter the num: "))
reversed_num=0
while num>0:
  digit=num%10
  reversed_num=reversed_num*10+digit
  num=num//10
print(reversed_num)
```

Q6. Count digits in a number.

Approach:
- Divide number by 10 repeatedly
- Count iterations

```python
num = int(input("Enter a number: "))
count = 0

if num == 0:
  count = 1
else:
  if num < 0:
    num = -num
  while num > 0:
    num = num // 10
    count = count +1

print(f"The number of digits is: {count}")
```

Q7. Check if a number is palindrome.

Approach:
- Reverse number
- Compare with original

```python
num = int(input("Enter a number: "))
original_num = num
reversed_num = 0

while num > 0:
  digit = num % 10
  reversed_num = reversed_num * 10 + digit
  num = num // 10

if original_num == reversed_num:
  print(f"{original_num} is a palindrome number")
else:
  print(f"{original_num} is not a palindrome number")
```

Q8. Guess the number game.

Approach:
- Use while loop
- Continue until correct guess

```python
#write answer here

original_num=10
num=int(input("Enter the Num: "))

while num!=original_num:
  if num>original_num:
    print("To high")
    num=int(input("Enter the Num: "))
  else:
    print("To low")
    num=int(input("Enter the Num: "))
print("Bingo Correct Guess")
```
